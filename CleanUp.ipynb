```python
import os
import shutil

filepath = 'test_files'
return_dict = {}
```

```python
fileSuffix_dict = {}
fileName_dict = {}

for _, _, files in os.walk(filepath):
    for file in files:
        if '.' in file:
            name, suffix = file.split('.')
        else:
            name, suffix = file, "others"
        
        fileSuffix_list = fileSuffix_dict.get(suffix, [])
        if not fileSuffix_list:
            fileSuffix_dict[suffix] = fileSuffix_list
        fileSuffix_list.append(file)

        fileName_list = fileName_dict.get(name, [])
        if not fileName_list:
            fileName_dict[name] = fileName_list
        fileName_list.append(file)
```

```python
remove_list = []
for k_name, v_name in fileName_dict.items():
    if len(v_name) > 1:
        for k_suffix, v_suffix in fileSuffix_dict.items():
            for v in v_suffix:
                if v in v_name:
                    v_suffix.remove(v)
            if len(v_suffix) == 0:
                remove_list.append(k_suffix)
            v_suffix.sort()
        for name in remove_list:
            del fileSuffix_dict[name]
        fileSuffix_dict[k_name] = v_name

fileSuffix_dict
```
